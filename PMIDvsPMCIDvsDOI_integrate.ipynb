```python
import math
def lookup_DOI(val, lookup, key="PMCID"): 
#     print(val, type(val))
#     if val is None or math.isnan(val):
#         return None
#     print(val)
    if type(val) == str:
        #print(val)
        #print(val, type(val))

        match = lookup[key] == val
        title = lookup['title'][match]
        try:
            return title.values[0]
        except:
            print("not found text")
            return None
    return None


df["title_2"] = df['text'].apply(lambda x: lookup_DOI(x, dfx, "sentence_text"))
```

```python
def title_or(row):
    t = None
    if type(row.title) == str:
        t=row.title
    elif type(row.title_2) == str:
        t=row.title_2
    else:
        t=row.Title
    return t
```
