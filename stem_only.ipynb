```python
import json
from octoflow.core import lemmatize
import pandas as pd

with open("downloads/sample_4k.json") as f:
    data = json.loads(f.read())
    
d = dict(data)
abst = [entry for entry in d["data"] if entry["abstract"]]
len(abst), abst[0]
```

```python
from pydash import flatten
```

```python
topics = [a["topics"] for a in abst]
journals = [a["journals"] for a in abst]
```

```python
STEM_stems = [
    "bio", #might be too permissive
    "phys",
    "energy",
    "engine",
    "metal",
    "surface",
    "protein",
    "medic",
    "tech",
    "compute",
    "artificial",
    "numer",
    "math",
    "chemi",
    "algorithm",
    "earth",
    "natural",
    "micro",
    "organ",
    "osis",
    "dynamics",
    "linear",
    "metabol",
    "physio",
    "electr",
    "equation",
    "graph",
    "scopy",
    "sensor",
    "agents",
    "informatic",
    "statist",
    "software",
    "meteor",
    "lipid",
    "machine",
    "vector",
    "metr",
    "icular",
    
    "neurosci",
    "dermat",
    "quantum",
    "urgy", #liturgy
    "voltage",
#    "clinical"
]
```

```python
with open("stem_stems.txt", "w") as f:
    f.write("\n".join(STEM_stems))
```

```python
STEM_absts = [a for a in abst if any(substring in " ".join(a['topics']).lower() for substring in STEM_stems)]
stem_topics = [a["topics"] for a in STEM_absts]

stem_terms = set(flatten(stem_topics))
all_terms = set(flatten(topics))
```
